# Plan de trabajo

**Descripción del objetivo** 

Al operador de telecomunicaciones Interconnect le gustaría poder pronosticar su tasa de cancelación de clientes. Si se descubre que un usuario o usuaria planea irse, se le ofrecerán códigos promocionales y opciones de planes especiales. El equipo de marketing de Interconnect ha recopilado algunos de los datos personales de sus clientes, incluyendo información sobre sus planes y contratos.  

Sus principales servicios son:
- Telefonía fija.
- Internet por línea telefónica o fibra óptica.
  También ofreve otros servicios como:
  - Sofware antivirus.
  - Soporte técnico.
  - Almacenamiento en la nube.
  - Streaming de TV y películas.

**Objetivo principal**: Crear un modelo predictivo de **clasificación** que arroje que clientes podrían cancelar sus servicios.  
(El cliente 'x' va a cancelar?, sí (1) o no (0)?)

**Pasos a realizar**  

1- Realizamos el preprocesamiento y análisis exploratotio de datos.  

 Parte 1:
- Identificar como están construidos nuestros datasets.
- Unificar datasets concatenando los 4 datasets por la columna 'customerID'.
- Limpiar y formatear los datos (minúsculas, espacios, reordenar y/o crear columnas...).
- Identificar y eliminar filas duplicadas.
- Analizar y resolver valores ausentes (eliminación, imputación...).
- Tratamiento de valores incongruentes.
- Cambiar tipo de datos de columnas.  

 Parte 2:
- Calcular estadísticas básicas (describe()).
- Crear visualizaciones (histogramas, gráficos de dispersión...).
- Identificar patrones (Buscar tendencias y relaciones entre variables).
- Crear hipótesis.

2- Preparación y entrenamiento del modelo (clasificación).
- En caso de existir un desbalance de clases, utilizar técnicas de balanceo (Class weight, sobremuestreo, submuestreo, ajuste de umbral...).
- Dividir dataset en entrenamiento (60%), validación (20%) y prueba (20%).
- Entrenar modelo de regresión logística y arbol de decisión como prueba de cordura.
- Entrenar modelos de boaque aleatorio y lightGBM.

3- Prueba de los modelos y elección del mejor
- Probar modelos.
- Mostrar métricas finales por modelo.
- Concluir y explicar elección del modelo.

4- Conclusiones generales del objetivo.
- Crear resumen de conslusiones y resultados del modelo predictivo diseñado (pueden contener gráficos).
